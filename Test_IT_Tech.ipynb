```python
# Производим объединение таблиц "PC", "Laptop", "Printer" с таблицей "Product" по ключу "maker"

table_pc = pd.merge(df_product, df_pc, on = 'model')
table_pc.head()
```

```python
table_laptop = pd.merge(df_product, df_laptop, on = 'model')
table_laptop.head()
```

```python
table_printer = pd.merge(df_product, df_printer, on = 'model')
table_printer.head()
```

```python
# Группируем данные в датафреймах по производителям и вычисляем для каждого из них среднюю цену по позиции "PC"

pc_mean_price_by_maker = table_pc.groupby('maker').agg({'price': 'mean'}).reset_index()
pc_mean_price_by_maker
```

```python
# Группируем данные в датафреймах по производителям и вычисляем для каждого из них среднюю цену по позиции "Laptop"

laptop_mean_price_by_maker_A = table_laptop.groupby('maker').agg({'price': 'mean'}).reset_index()
laptop_mean_price_by_maker_A
```

```python
# Группируем данные в датафреймах по производителям и вычисляем для каждого из них среднюю цену по позиции "Printer"

printer_mean_price_by_maker = table_printer.groupby('maker').agg({'price': 'mean'}).reset_index()
printer_mean_price_by_maker
```

```python
# Определяем среднюю цену для позиций "PC" & "Laptop" & "Printer" для компаний производителей "A" или "D"
pc_mean_price_by_maker_A = pc_mean_price_by_maker.query('maker == "A"')
laptop_mean_price_by_maker_A = laptop_mean_price_by_maker.query('maker == "A"')
printer_mean_price_by_maker_A = printer_mean_price_by_maker.query('maker == "A"')
printer_mean_price_by_maker_D = printer_mean_price_by_maker.query('maker == "D"')
mean_price_by_maker_A_D = [pc_mean_price_by_maker_A.price, laptop_mean_price_by_maker_A.price, printer_mean_price_by_maker_A.price, printer_mean_price_by_maker_D.price]
```

```python
# Выводим общую среднюю цену для всех "PC" & "Laptop" & "Printer" для компаний производителей "A" & "D"

round(np.mean(mean_price_by_maker_A_D), 1)
```
